```python
def outlier_remover(s):
    q1 = np.percentile(s,25)
    q3 = np.percentile(s,75)
    iqr = q3-q1
    lf  = q1 - 1.5*iqr
    uf  = q3 + 1.5*iqr  
    return lf, uf
```

```python
lf,uf = outlier_remover(data.Salary)
```

```python
data[(data.Salary<lf) | (data.Salary>uf)]
```
