# Adding a col and  generat values on conditions

#### list comprehension 

```python
emp['Bonus'] = [50000 if age>=60
                else 40000 if age>=50 
                else 30000 if age>=40 
                else 20000 if age>=30 
                else 10000 if age>=20 
                else 5000 
                for age in emp.Age]
emp
```

```python
emp['Bonus1']=[sal*.50 if sal>=200000 
               else sal*.30 if sal>=150000
               else sal*.25 if sal>=100000 
               else sal*.15 if sal>=75000 
               else sal*.10  
               for sal in emp.Salary]
emp
```

# using function

```python
def bonus(sal):
    if sal>=200000:
        return sal*.50
    elif sal>=150000:
        return sal*.30
    elif sal>=100000:
        return sal*.25
    elif sal>=75000:
        return sal*.15
    else:
        return sal*.10  
```

```python
emp['bonus2'] = emp.Salary.apply(bonus)
emp
```

```python
emp['bonus3']  = emp.Salary.apply(lambda sal:sal*.50 if sal>=200000 
                                               else sal*.30 if sal>=150000
                                               else sal*.25 if sal>=100000 
                                               else sal*.15 if sal>=75000 
                                               else sal*.10 )
```
